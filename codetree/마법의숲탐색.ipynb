```python
def move_golem(in_y, di):
    x, y = 1, in_y  # 골렘이 입장 전 상태
    global board
    # 더이상 움직일 수 없을 때 까지
    while True:
        # 남쪽 이동
        # 왼아, 가아, 오아 빈칸
        if sum([board[x+1][y-1], board[x+2][y], board[x+1][y+1]]) == 0:
            x += 1  # 남쪽 이동

        # 서쪽 이동 -> 남쪽 이동
        # 왼위, 왼가, 왼아, 왼왼아, 왼아아 빈칸
        elif sum([board[x-1][y-1], board[x][y-2], board[x+1][y-1], board[x+1][y-2], board[x+2][y-1]]) == 0:
            y -= 1  # 서쪽 이동
            x += 1  # 남쪽 이동
            di = (di - 1) % 4   # 출구 반시계 이동

        # 동쪽 이동 -> 남쪽 이동
        # 오위, 오가, 오아, 오오아, 오아아 빈칸
        elif sum([board[x-1][y+1], board[x][y+2], board[x+1][y+1], board[x+1][y+2], board[x+2][y+1]]) == 0:
            y += 1  # 동쪽 이동
            x += 1  # 남쪽 이동
            di = (di + 1) % 4   # 출구 시계 이동

        # 더이상 이동 불가능
        else:
            break

    return x, y, di

# 골렘이 튀어나갔는지 확인
def is_board(x):
    # x를 늘린 것을 고려해야함 / 골렘의 가운데가 4에 와야 골렘이 모두 안에 위치
    return x < 4

# 골렘 격자에 위치 시키기
def set_golem(x, y, di):
    global cnt, exits

    board[x][y-1:y+2] = [cnt] * 3
    board[x-1][y] = board[x+1][y] = cnt

    cnt += 1    # 다음 골렘
    exits.append([x + dx[di], y + dy[di]])  # 골렘의 출구 저장

from collections import deque
# 정령 이동
def move_elf(x, y):
    visited = [[0]*(c+2) for _ in range(r+4)]

    q = deque()
    q.append([x, y])
    visited[x][y] = 1
    max_x = 0

    while q:
        x, y = q.popleft()
        max_x = max(x, max_x)
        for d in range(4):
            nx = x + dx[d]
            ny = y + dy[d]

            # 이미 방문함
            if visited[nx][ny] == 1: continue
            # 골렘 안에서 가장 남쪽으로 움직임 / 골렘의 출구로 나가서 인접한 골렘으로 이동
            if (board[nx][ny] == board[x][y]) or ((board[nx][ny] > 1) and [x, y] in exits):
                q.append([nx, ny])
                visited[nx][ny] = 1

    return max_x - 2    # 골렘 입장을 위해 늘린만큼 제거/격자 1로 시작
```

```python
def check_(x, y):
    if point_matrix_check(x, y, matrix):
        return False
    if matrix[x][y] != 0:
        return False
    else:
        return True

def check(x, y):
    if check_(x+1,y-1) and check_(x+2,y) and check_(x+1,y+1):
        return 'down'
    elif check_(x, y-2) and check_(x-1, y-1) and check_(x+1, y-1) and check_(x+1,y-2) and check_(x+2, y-1):
        return 'left_down'
    elif check_(x-1,y+1) and check_(x,y+2) and check_(x+1,y+1) and check_(x+1,y+2) and check_(x+2,y+1):
        return 'right_down'
    else:
        return 'no'

def point_matrix_check(x, y, matrix):
    if x < 0 or x >= len(matrix) or y < 0 or y >= len(matrix[0]):
        return True
    else:
        return False
        
def matrix_check(x, y, matrix):
    if x < 0 or x >= len(matrix) - 1 or y < 1 or y >= len(matrix[0]) - 1:
        return True
    else:
        return False

def print_matrix(center, idx, prev_center=None):
    global matrix
    if prev_center:
        matrix[prev_center[0]][prev_center[1]] = 0
        for dx, dy in zip([-1,0,1,0],[0,-1,0,1]):
            matrix[prev_center[0]+dx][prev_center[1]+dy] = 0
    matrix[center[0]][center[1]] = idx+1
    for dx, dy in zip([-1,0,1,0],[0,-1,0,1]):
        if point_matrix_check(center[0]+dx, center[1]+dy, matrix):
            continue
        matrix[center[0]+dx][center[1]+dy] = idx+1

def gol_move(gol_dict, idx):
    global exit_list
    curr_center = gol_dict[idx]['center']
    # print_matrix(curr_center, idx)
    while True:
        if matrix_check(curr_center[0], curr_center[1], matrix):
            break
        check_down = check(curr_center[0], curr_center[1])
        curr_center_ = curr_center
        if check_down == 'down':            
            curr_center = [curr_center[0]+1, curr_center[1]]
        elif check_down == 'left_down':
            curr_center = [curr_center[0]+1, curr_center[1]-1]
            gol_dict[idx]['exit'] = exit_list[exit_list.index(gol_dict[idx]['exit']) - 1]
        elif check_down == 'right_down':
            curr_center = [curr_center[0]+1, curr_center[1]+1]
            idx_ = exit_list.index(gol_dict[idx]['exit']) + 1
            if idx_ >= len(exit_list):
                idx_ = 0
            gol_dict[idx]['exit'] = exit_list[idx_]
        elif check_down == 'no':
            break
        print_matrix(curr_center, idx, curr_center_)
    gol_dict[idx]['center'] = [curr_center[0], curr_center[1]]
    return gol_dict

def move(gol_dict, idx):
    curr_center = gol_dict[idx]['center']
    visited = [[False] * len(matrix[0]) for _ in range(len(matrix))]
    # visited[0:3] = [[True]*len(matrix[0])] * 3
    visited[0] = [True] * len(matrix[0])
    queue = []
    queue.append([curr_center[0], curr_center[1]])
    visited[curr_center[0]][curr_center[1]] = True
    curr_row = curr_center[0]
    while queue:
        row, col = queue.pop()
        if idx == 3:
            print(row, col)
        # if row == 7 and col == 3:
        #     print('found')
        if visited[row][col]:
            continue        
        
        for dx, dy in zip([-1,0,1,0], [0,-1,0,1]):
            if point_matrix_check(row+dx, col+dy, matrix):
                continue
            if matrix[row+dx][col+dy] == 0:
                continue
            gol_exit = gol_dict[matrix[row][col]-1]['exit']
            gol_center = gol_dict[matrix[row][col]-1]['center']
            curr_exit = [gol_center[0] + gol_exit[0], gol_center[1] + gol_exit[1]]
            if matrix[row+dx][col+dy] != matrix[row][col]:
                next_gol_exit = gol_dict[matrix[row+dx][col+dy]-1]['exit']
                next_gol_center = gol_dict[matrix[row+dx][col+dy]-1]['center']
                next_exit = [next_gol_center[0] + next_gol_exit[0], next_gol_center[1] + next_gol_exit[1]]
                if (curr_exit[0] == row and curr_exit[1] == col) and (next_exit[0] != row + dx or next_exit[1] != col + dy):
                    queue.append([row+dx, col+dy])
                    visited[row][col] = True
            elif matrix[row+dx][col+dy] == matrix[row][col]:
                queue.append([row+dx, col+dy])
                visited[row][col] = True
        if curr_row < row:
            curr_row = row
    return curr_row
```
