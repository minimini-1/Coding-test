```python
from copy import deepcopy

K, m = map(int, input().split())
matrix = [[0]] * 5
m_list = []
for _ in range(5):
    matrix[_] = list(map(int, input().split()))
m_list = list(map(int, input().split()))

dx = [1, 0, -1, 0]
dy = [0, 1, 0, -1]
new_index = 0

def calculate_value(board):
    total_remove_list = []
    value = 0
    visited2 = [[False] * 5 for i in range(5)]
    for i in range(5):
        for j in range(5):
            if visited2[i][j]:
                continue
            stack = []
            target = board[i][j]
            stack.append([i, j])
            visited = [[False] * 5 for i in range(5)]
            cnt = 0
            remove_list = []

            while stack:
                x, y = stack.pop()
                if x < 0 or x >= 5 or y < 0 or y >= 5:
                    continue
                if visited[x][y]:
                    continue
                visited[x][y] = True

                if board[x][y] == target:
                    cnt += 1
                    remove_list.append([x, y])
                    for k in range(4):
                        nx = x + dx[k]
                        ny = y + dy[k]
                        stack.append([nx, ny])
            if cnt >= 3:
                value += cnt
                total_remove_list.extend(remove_list)
                for x, y in remove_list:
                    visited2[x][y] = True
    return value, total_remove_list

def rotate_site(x, y, degree):
    copy_matrix = deepcopy(matrix)
    for _ in range(degree):
        temp = copy_matrix[x-1][y]
        copy_matrix[x-1][y] = copy_matrix[x][y-1]
        copy_matrix[x][y-1] = copy_matrix[x+1][y]
        copy_matrix[x+1][y] = copy_matrix[x][y+1]
        copy_matrix[x][y+1] = temp

        temp = copy_matrix[x-1][y-1]
        copy_matrix[x-1][y-1] = copy_matrix[x+1][y-1]
        copy_matrix[x+1][y-1] = copy_matrix[x+1][y+1]
        copy_matrix[x+1][y+1] = copy_matrix[x-1][y+1]
        copy_matrix[x-1][y+1] = temp
    return calculate_value(copy_matrix)

def rotate_real_site(x, y, degree):
    for _ in range(degree):
        temp = matrix[x-1][y]
        matrix[x-1][y] = matrix[x][y-1]
        matrix[x][y-1] = matrix[x+1][y]
        matrix[x+1][y] = matrix[x][y+1]
        matrix[x][y+1] = temp

        temp = matrix[x-1][y-1]
        matrix[x-1][y-1] = matrix[x+1][y-1]
        matrix[x+1][y-1] = matrix[x+1][y+1]
        matrix[x+1][y+1] = matrix[x-1][y+1]
        matrix[x-1][y+1] = temp

def fill_site(rl):
    global new_index, matrix
    rl.sort(key=lambda x: (x[1], -x[0]))
    for x, y in rl:
        matrix[x][y] = m_list[new_index]
        new_index += 1

def exploration():
    max_value = 0
    max_index_degree = []
    remove_list = []

    for k in range(1,4):
        for i in range(1,4):
            for j in range(1,4):
                v, r_l = rotate_site(i, j, k)
                if max_value < v:
                    max_value = v
                    max_index_degree = [i, j, k]
                    remove_list = r_l
    
    if max_value == 0:
        return max_value
    
    rotate_real_site(max_index_degree[0], max_index_degree[1], max_index_degree[2])
    fill_site(remove_list)

    v, rl = calculate_value(matrix)
    while v != 0:
        fill_site(rl)
        max_value += v
        v, rl = calculate_value(matrix)

    return max_value


result = []
for _ in range(K):
    value = exploration()
    if value == 0:
        break
    result.append(value)
print(" ".join(map(str, result)))
```
